```python
## Question 1: Use yfinance to exctract stock data

## Reset the index, save, and display the first five rows of the tesla_data DataFrame using the head function.

import yfinance as yf

tesla_data = yf.Ticker("TSLA")
tesla_data = tesla_data.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
## Question 2: Use WebScraping to etract Tesla Revenue Data
## Display the last 5 rows of the tesla_revenue DataFrame using the tail function.

import pandas as pd

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'

dataframes = pd.read_html(url, match='Tesla Annual Revenue', flavor='bs4')

tesla_revenue = dataframes[0]

tesla_revenue.tail()


```

```python
# Question 3: Use yfinance to extract stock data
# Reset the index, save, and display the first five rows of the GameStop stock data DataFrame using the head function.

gamestop_data = yf.Ticker("GME")
gamestop_data = gamestop_data.history(period="max")

gamestop_data.reset_index(inplace=True)
gamestop_data.head()
```

```python
# Question 4: Use WebScraping to extract GameStop Revenue Data
# Display the last five rows of the GameStop revenue data DataFrame using the tail function.
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'
html_data_2 = requests.get(url).text
soup = BeautifulSoup(html_data_2, 'html.parser')

# Find all tables and get index of GameStop Annual Revenue
tables = soup.find_all('table')
for index, table in enumerate(tables):
    if ("GameStop Annual Revenue" in str(table)):
        table_index = index

gamestop_revenue = pd.DataFrame(columns=["Year", "Revenue"])
revenues_and_years = []

# Get the data and append it to the DataFrame
for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        year = col[0].text
        revenue = col[1].text
        revenues_and_years.append({"Year": year, "Revenue": revenue})

gamestop_revenue = pd.DataFrame(revenues_and_years)
gamestop_revenue.tail()
```
